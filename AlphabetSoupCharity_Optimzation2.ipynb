## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
#import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
AlphabetSoupCharity_Optimzation = pd.read_csv("charity_data.csv")
AlphabetSoupCharity_Optimzation.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
AlphabetSoupCharity_Optimzation=AlphabetSoupCharity_Optimzation.drop(columns=['EIN', 'NAME','AFFILIATION'])
#charity=charity.drop(labels='EIN',axis=0)
AlphabetSoupCharity_Optimzation.head()
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
AlphabetSoupCharity_Optimzation.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
#AlphabetSoupCharity_Optimzation.groupby('APPLICATION_TYPE').count()
app_types=AlphabetSoupCharity_Optimzation['APPLICATION_TYPE'].value_counts()
app_types
```

```python
application_types_to_replace =app_types[app_types <500]
application_types_to_replace=application_types_to_replace.index.values
```

```python
application_types_to_replace
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
application_types_to_replace =app_types[app_types <50]
application_types_to_replace=application_types_to_replace.index.values
# Replace in dataframe
for app in application_types_to_replace:
    AlphabetSoupCharity_Optimzation['APPLICATION_TYPE'] = AlphabetSoupCharity_Optimzation['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
AlphabetSoupCharity_Optimzation['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
#AlphabetSoupCharity_Optimzation.groupby('CLASSIFICATION').count()
class_types=AlphabetSoupCharity_Optimzation['CLASSIFICATION'].value_counts()
class_types
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
class_types[class_types >1]
```

```python
 classifications_to_replace =class_types[class_types <1000]
 classifications_to_replace=classifications_to_replace.index.values
classifications_to_replace
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
#classifications_to_replace =class_types[class_types <100]
#classifications_to_replace=classifications_to_replace.index.values

# Replace in dataframe
for cls in classifications_to_replace:
    AlphabetSoupCharity_Optimzation['CLASSIFICATION'] = AlphabetSoupCharity_Optimzation['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
AlphabetSoupCharity_Optimzation['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
application_dummies = pd.get_dummies(AlphabetSoupCharity_Optimzation)
application_dummies
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = application_dummies["IS_SUCCESSFUL"].values
X = application_dummies.drop(["IS_SUCCESSFUL"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
no_input_features = X_train_scaled.shape[1]
no_input_features
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()
nn_model = tf.keras.models.Sequential()
# First hidden layer and input layer
nn_model.add(tf.keras.layers.Dense(units=99, activation="relu", input_dim=no_input_features))
# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=69, activation="relu"))
# third layer
nn_model.add(tf.keras.layers.Dense(units=30, activation="relu"))

# Output layer

nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# nn_model.build()
# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE 
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=50) #, callbacks=)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
